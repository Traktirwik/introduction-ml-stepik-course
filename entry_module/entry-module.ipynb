```python
my_stat = pd.read_csv('./models/my_stat.csv')
my_stat.head()
```

```python
my_stat = my_stat.rename(columns= {'V1': 'session_value', 'V2': 'group', 'V3': 'time', 'V4': 'n_users'})
my_stat.head()
```

```python
my_stat_1 = pd.read_csv('./models/my_stat_1.csv')
```

```python
my_stat_1 = my_stat_1.fillna({'session_value': 0})
getNegativeValues = my_stat_1.query('n_users >= 0')
med = getNegativeValues['n_users'].median()
my_stat_1.loc[my_stat_1['n_users'] < 0] = med
my_stat_1.head()
```

```python
my_stat_1['session_value'] = my_stat_1['session_value'].fillna(0)
my_stat_1.loc[my_stat["n_users"] < 0, "n_users"] = \
    my_stat_1.loc[my_stat["n_users"] >= 0, "n_users"].median()
my_stat_1.head()
```

```python
mean_session_value_data = my_stat_1.groupby('group', as_index=False).agg({'session_value': 'mean'}).rename(columns={'session_value': 'mean_session_value'})
mean_session_value_data
```
